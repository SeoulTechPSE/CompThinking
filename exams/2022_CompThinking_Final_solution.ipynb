$~$Write a Python programming to create a pie chart with a title of **The most popular programming languages in 2022**


> The code snippet gives the output shown in the following screenshot:

![2022_Final_1](https://raw.githubusercontent.com/SeoulTechPSE/CompThinking/main/exams/figs/2022_Final_1.png)

```python
import matplotlib.pyplot as plt

# Plot data
languages = 'Python', 'C', 'Java', 'C++', 'C#', 'Visual Basic', 'JavaScript', 'etc'
popuratity = [12.74, 11.59, 10.99, 8.83, 6.39, 5.86, 2.12, 41.48]

colors = ['red', 'gold', 'yellowgreen', 'turquoise', 'lightcoral', 'lightskyblue', 'beige', 'silver']

# explode 1st slice
explode = (0.1, 0, 0, 0, 0, 0, 0.1, 0)  

fig, ax = plt.subplots(figsize=(8, 8))

# Plot
ax.pie(popuratity, explode=explode, labels=languages, colors=colors, autopct='%1.1f%%', startangle=45)
plt.title("The most popular programming languages in 2022")

plt.savefig('2022_Final_1.png');
```

#### 위와 같은 `DataFrame`을 만들고 또한 `pivot_table` 함수를 사용하여 아래와 같은 피벗 테이블과  `heatmap`을 만드는 코드를 작성하시오.

```python
import pandas as pd

data = [['Oranges', 'North', 12.30],
       ['Apples', 'South', 10.55],
       ['Oranges', 'South', 22.00],
       ['Bannas', 'South', 30.90],
       ['Bannas', 'North', 5.30],
       ['Oranges', 'North', 31.10],
       ['Apples', 'North', 40.12]]

sales = pd.DataFrame(data=data, columns=['Fruit', 'Region', 'Revenue'])
sales
```

```python
pivot = pd.pivot_table(sales, 
                       index='Fruit', columns='Region', values='Revenue', aggfunc='sum',
                       margins=True, margins_name='Total')
pivot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
sns.heatmap(pivot.iloc[:3, :2], ax=ax, linewidths=.5);
```
